```python
#Player stats per game 
year=2021
data_scrap = []
for i in team_list:
    team_url = (f'https://www.basketball-reference.com/teams/{i}/{str(year)}.html')
    team_response = requests.get(team_url)
    team_soup = BeautifulSoup(team_response.content, 'lxml')
    
    
    per_game = team_soup.find(name = 'table', attrs = {'id' : 'per_game'})

    for row in per_game.find_all('tr')[1:]: 

        player = {}
        player['Name'] = row.find('a').text.strip()
               
        player['Team'] = i
            
        player['Age'] = row.find('td', {'data-stat' : 'age'}).text
        player['Games'] = row.find('td', {'data-stat' : 'g'}).text
        player['Games_Start'] = row.find('td', {'data-stat' : 'gs'}).text
        player['Min_PG'] = row.find('td', {'data-stat' : 'mp_per_g'}).text
        
        player['Field_Goals'] = row.find('td', {'data-stat' : 'fg_per_g'}).text
        player['Field_Goals_Attempts'] = row.find('td', {'data-stat' : 'fga_per_g'}).text
        
        player['3Point_Goals'] = row.find('td', {'data-stat' : 'fg3_per_g'}).text
        player['3Point_Goals_Attempts'] = row.find('td', {'data-stat' : 'fg3a_per_g'}).text
        
        player['2Point_Goals'] = row.find('td', {'data-stat' : 'fg2_per_g'}).text
        player['2Point_Goals_Attempts'] = row.find('td', {'data-stat' : 'fg2a_per_g'}).text
        
        player['Effective_Field_Goal_Per'] = row.find('td', {'data-stat' : 'efg_pct'}).text
        
        player['Free_Throw'] = row.find('td', {'data-stat' : 'ft_per_g'}).text
        player['Free_Throw_Attempts'] = row.find('td', {'data-stat' : 'ft_per_g'}).text
        
        player['Offensive_Rebounds'] = row.find('td', {'data-stat' : 'orb_per_g'}).text
        player['Defensive_Rebounds'] = row.find('td', {'data-stat' : 'drb_per_g'}).text
        
        player['Assists'] = row.find('td', {'data-stat' : 'ast_per_g'}).text
        player['Steals'] = row.find('td', {'data-stat' : 'stl_per_g'}).text
        player['Blocks'] = row.find('td', {'data-stat' : 'blk_per_g'}).text
        player['Turnovers'] = row.find('td', {'data-stat' : 'tov_per_g'}).text
        
        player['Points'] = row.find('td', {'data-stat' : 'pts_per_g'}).text

        

    
        #Player Position information                              
        player_url = ('https://www.basketball-reference.com/' + row.find('a').attrs['href'])
        player_response = requests.get(player_url)
        player_soup = BeautifulSoup(player_response.content, 'lxml')
        player_info = player_soup.find(name = 'div', attrs = {'itemtype' : 'https://schema.org/Person'})

       

        s = str(player_info.find_all('p'))

            
        position = re.search('Position:\n  </strong>\n (.*)\n\n', s)
            
        player['Position'] = position.group(1).strip()

        data_scrap.append(player)
        
        time.sleep(3)
        
nba_df = pd.DataFrame(data_scrap)

```
